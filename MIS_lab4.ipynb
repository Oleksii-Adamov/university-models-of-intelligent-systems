# Рівняння

$4.2x_1+10x_2=136$

$-10.1x_1+13.8x_2=132$

$18.3x_1-7.6x_2=108$

# Графіки

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def line1(x):
    return (136 - 4.2*x) / 10

def line2(x):
    return (132 + 10.1*x) / 13.8

def line3(x):
    return (108 - 18.3*x) / (-7.6)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 5))
x1 = np.linspace(0, 20, 400)
for i in range(0, 3):
    axes[i].set(xlim = (0, 20), ylim = (-15, 30), xlabel = '$x_1$', ylabel = '$x_2$')
axes[0].set_title('$4.2x_1+10x_2=136$')
axes[0].plot(x1, line1(x1), '--', c=np.random.rand(3))
axes[1].set_title('$-10.1x_1+13.8x_2=132$')
axes[1].plot(x1, line2(x1), '-.', c=np.random.rand(3))
axes[2].set_title('$18.3x_1-7.6x_2=108$')
axes[2].plot(x1, line3(x1),':',c=np.random.rand(3))
fig.tight_layout()
plt.savefig('mis_lab4_1.jpg')
plt.savefig('mis_lab4_1.png')
plt.savefig('mis_lab4_1.svg')
plt.show()
```

```python
int_1_2 = np.linalg.solve(np.array([[4.2, 10], [-10.1, 13.8]]), np.array([136, 132]))
int_1_2
```

```python
int_1_3 = np.linalg.solve(np.array([[4.2, 10], [18.3, -7.6]]), np.array([136, 108]))
int_1_3
```

```python
int_2_3 = np.linalg.solve(np.array([[-10.1, 13.8], [18.3, -7.6]]), np.array([132, 108]))
int_2_3
```

```python
x1 = np.linspace(0, 20, 400)

plt.figure(figsize=(10, 6), dpi=100)

plt.plot(x1, line1(x1), color=np.random.rand(3), linestyle='dotted', label='$f[0]=4.2x_1+10x_2=136$')
plt.plot(x1, line2(x1), color=np.random.rand(3), linestyle='dashed', label='$f[1]=-10.1x_1+13.8x_2=132$')
plt.plot(x1, line3(x1), color=np.random.rand(3), linestyle='dashdot', label='$f[2]=18.3x_1-7.6x_2=108$')

plt.xlim(0, 20)
plt.ylim(0, 25)

# Додавання підписів та заголовка
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('All lines in one figure')

plt.text(12.5, line1(12.5), '$4.2x_1+10x_2=136$', ha='left', va='bottom',
         transform_rotates_text=True, rotation=np.degrees(np.arctan(-0.42)), rotation_mode='anchor')
plt.text(5, line2(5), '$-10.1x_1+13.8x_2=132$', ha='left', va='bottom',
         transform_rotates_text=True, rotation=np.degrees(np.arctan(10.1 / 13.8)), rotation_mode='anchor')
plt.text(10.5, line3(10.5), '$18.3x_1-7.6x_2=108$', ha='left', va='bottom',
         transform_rotates_text=True, rotation=np.degrees(np.arctan(18.3 / 7.6)), rotation_mode='anchor')

plt.legend()

plt.grid(color='gray', linestyle='-')

# Заповнення області перетину
intersection_x_space = np.linspace(3.50276799, 14.18591421, 400)
plt.fill_between(intersection_x_space, line2(intersection_x_space), np.concatenate((line1(intersection_x_space[intersection_x_space < 9.83435697]), line3(intersection_x_space[intersection_x_space > 9.83435697]))), color='gray', alpha=0.3)

# Позначення точок перетину
plt.scatter(int_1_2[0], int_1_2[1], color='red', s=100, marker='o', linewidth=2)
plt.annotate(f'The intersection point of f[0] and f[1]', xy=int_1_2, xytext=(int_1_2[0] - 3.4, int_1_2[1] + 5),
                 arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.scatter(int_1_3[0], int_1_3[1], color='blue', s=100, marker='o', linewidth=2)
plt.annotate(f'The intersection point of f[0] and f[2]', xy=int_1_3, xytext=(int_1_3[0] - 3.4, int_1_3[1] - 5),
                 arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.scatter(int_2_3[0], int_2_3[1], color='green', s=100, marker='o', linewidth=2)
plt.annotate(f'The intersection point of f[1] and f[2]', xy=int_2_3, xytext=(int_2_3[0] - 1, int_2_3[1] - 5),
                 arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.savefig('mis_lab4_2.jpg')
plt.savefig('mis_lab4_2.png')
plt.savefig('mis_lab4_2.svg')

plt.show()
```
